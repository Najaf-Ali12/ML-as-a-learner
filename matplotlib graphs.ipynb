```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=np.arange(0,10)
y=np.arange(31,41)
plt.scatter(x,y,c='r')
plt.xlabel('height')
plt.ylabel('weight')
plt.title('Graph in 2D')
```

```python
y=x*x
plt.plot(x,y,'b*')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Understanding plot")
```

```python
plt.subplot(2,2,1)  #here as your increase the 10 the HEIGHT decreases and as increase the width decreases and 1 shows position of subplot
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
```
