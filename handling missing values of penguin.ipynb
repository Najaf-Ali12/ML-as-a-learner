```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=sns.load_dataset("Penguins")
```

```python
data.info()
```

```python
data.isnull().sum().sort_values(ascending=False)  #view columns with missing values
```

```python
#Visualizing the missing values
plt.figure(figsize=(8,5))
sns.heatmap(data.isnull(), cbar=False)
plt.show()                    
```

```python
data["sex"]=data["sex"].fillna(data["sex"].mode()[0])
```

```python
data["bill_length_mm"]=data["bill_length_mm"].fillna(data["bill_length_mm"].mean())           
```

```python
data["bill_depth_mm"]=data["bill_depth_mm"].fillna(data["bill_depth_mm"].median())   
```

```python
data.drop("body_mass_g", axis=1, inplace=True)
```

```python
data.drop("flipper_length_mm", axis=1, inplace=True)
```
