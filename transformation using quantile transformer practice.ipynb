```python
#impoting required libraries
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import QuantileTransformer
```

```python
#creating non-gaussian distribution
data=np.random.exponential(size=1000, scale=2)
df=pd.DataFrame(data, columns=["Values"])
df.head()
```

```python
#histogram of data
print("Diagram before transformation")
sns.histplot(df["Values"], kde=1)
```

```python
#TRANSFORMING PROCESS
qtl_normal=QuantileTransformer(output_distribution="normal")
df["Quantile"]=qtl_normal.fit_transform(df[["Values"]])
df["Quantile"].head()
```

```python
#histogram of normal distribution
print("Histogram after quantile distribution")
sns.histplot(df["Quantile"], kde=1)
```
