TRANSFORMATION:
Transformation is also called as Normalization . In actual transformation is normalization


A Gaussian distribution or normal distribution is a bell-shaped curve where most data points are around the average, with fewer data points as you move away from the average. It's used to describe things like people's heights or test scores.

In this program we will transform non gaussian distribution into gaussian or normal distribution

```python
#importing required libraries
import pandas as pd
import numpy as np 
#import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#generating non-normal data(exponential data)
np.random.seed(0)
df=np.random.exponential(size=1000, scale=2)
df=pd.DataFrame(df, columns=["values"])
print(df.shape)  #it will output(no_of_rows, no_of_columns)
df.head()
```

```python
#creating the histogram of the non-normal data
sns.histplot(df["values"], kde=True)
```

```python
#importing requirements for transformation of data
from sklearn.preprocessing import PowerTransformer
#we will divide the powertransformer into two methods
pt_boxcox=PowerTransformer(method="box-cox")
pt_yeo_johnson=PowerTransformer(method="yeo-johnson")
```

```python
#applying box-cox transformer
#data must be positive for boxcox transformer method
df["Box_Cox"]=pt_boxcox.fit_transform(df[["values"]])
```

```python
#visualing the histogram of transformed data through box_cox method
sns.histplot(df["Box_Cox"], kde=True)
```

```python
#applying Yoe_Johnson transformer method
df["Yeo_johnson"]=pt_yeo_johnson.fit_transform(df[["values"]])
#visualizing the histogram of transformed data obtained using yeo_johnson method by sns.histplot method
sns.histplot(df["Yeo_johnson"], kde=1)
```

```python
#quantile transformer
from sklearn.preprocessing import QuantileTransformer
qtl_normal=QuantileTransformer(output_distribution="normal")
df["Quantile"]=qtl_normal.fit_transform(df[["values"]])
df["Quantile"]
```

```python
#visualizing quantile transformer result using histograms
sns.histplot(df["Quantile"], kde=1)
```
