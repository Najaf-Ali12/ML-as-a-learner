```python
#importing required libraries
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
#creation of skewed data (exponential and lognormal, types of data distribution)
data={
    "Column1":np.random.exponential(size=100, scale=2),
    "Column2":np.random.exponential(size=100, scale=3)}
```

```python
#creating histogram of the non-normal or skewed data
print("The histogram diagram of first column of non-gaussian distributed data")
sns.histplot(data["Column1"], kde=True)    #kde will create a line
```

```python
print("The histogram diagram of second column of non-gaussian distributed data")
sns.histplot(data["Column2"], kde=True)
```

```python
#converting data in dataframe
data=pd.DataFrame(data)
data.shape   #it will give output(rows,column)
```

```python
from sklearn.preprocessing import PowerTransformer  
pt_box_cox=PowerTransformer(method="box-cox")
pt_yeo_johnson=PowerTransformer(method="yeo-johnson")
#here we will transform data using two methods of powertransformer
```

```python
#transformation of first column of given data
data["box-cox"]=pt_box_cox.fit_transform(data[["Column1"]])  # a new column box-cox is added in dataset
data["box-cox"]
```

```python
print("The histogram of first column of non-normal data after transforming in normal distribution")
sns.histplot(data["box-cox"], kde=1)
```

```python
#transformation of second column of given data in guassian distribution
data["Yeo-Johnson"]=pt_yeo_johnson.fit_transform(data[["Column2"]])  #a new column named yeo_johnson is added
data["Yeo-Johnson"]
```

```python
print("The histogram of second column of non-normal data after transforming in normal distribution")
sns.histplot(data[["Yeo-Johnson"]], kde=1)
```
